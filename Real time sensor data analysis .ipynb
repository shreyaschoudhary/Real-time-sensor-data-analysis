```
#Importing the necessary libraries
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import statistics as st
import seaborn as sns
```

```
#Reading the data into a dataframe
data = pd.read_csv('https://raw.githubusercontent.com/shreyaschoudhary/Process-Data-Analytics/master/Data4T_expt_clean.csv')
```

```
#Displaying the given data in the dataframe
data
```

-----------------------Part (A)---------------------------------------

```
#Initializing variables to the volts data for cold, hot and total sensors
coldv = data['Cold (Volts)']
hotv = data['Hot (Volts)']
totalv = data['Total (Volts)']
```

```
#Converting to Mass flow rates based on the emperical equation provided in the document
data['mcw'] = (1.256 + (178.032*(np.sqrt(coldv))))
data['mhw'] = (1.444 + (181.574*(np.sqrt(hotv))))
data['mt'] = (-2.319 + (355.822*(np.sqrt(totalv))))
```

```
#Adding the mass flow rates to the dataframe and printing it
data
```

------------------------Part (B)-------------------------------

```
#Plotting the mass flow rates of cold water, hot water and total with respect to time
t = data['Time (Sec.)']
cw= data['mcw']
hw = data['mhw']
tot = data['mt']
plt.figure(figsize=(20,10)) 
plt.plot(t, cw, label = 'Cold Water')                                     #Plotting Mass flow rate of cold water v/s time
plt.plot(t,hw,label = 'Hot Water')                                        #Plotting Mass flow rate of hot water v/s time
plt.plot(t,tot,label = 'Total')                                           #Plotting total Mass flow rate v/s time
plt.xlabel('Time (Sec.)', fontsize=12)
plt.ylabel('Mass flow rates of cw, hw and total (kg/h)',fontsize=12)
plt.title('Mass flow rates V/s Time plot',fontsize=18)
plt.legend()
```

```
#Splitting the flow rates data based on the plot above
cw1 = cw.iloc[0:1500]           #Split 1 for cold water flow rate
cw2 = cw.iloc[1990:]            #Split 2 for cold water flow rate
hw1= hw.iloc[0:1500]            #Split 1 for hot water flow rate
hw2= hw.iloc[1990:]             #Split 2 for hot water flow rate
tot1= tot.iloc[0:1500]          #Split 1 for total flow rate
tot2 = tot.iloc[1990:]          #Split 2 for total flow rate
```

```
#Calculating mean of the split data for each flow rates

#For cold water mass flow rate
meancw1 = st.mean(cw1)
meancw2 = st.mean(cw2)

#For hot water mass flow rate
meanhw1 = st.mean(hw1)
meanhw2 = st.mean(hw2)

#For total mass flow rate
meantot1 = st.mean(tot1)
meantot2 = st.mean(tot2)

#Printing the values
print("The mean of first split of data for flow rate of cold water (kg/h), containing 1500 points is:", meancw1)
print("The mean of second split of data for flow rate of cold water (kg/h), containing 1000 points is:", meancw2)

print("The mean of first split of data for flow rate of hot water (kg/h), containing 1500 points is:", meanhw1)
print("The mean of second split of data for flow rate of hot water (kg/h), containing 1000 points is:", meanhw2)

print("The mean of first split of data for total flow rate (kg/h), containing 1500 points is:", meantot1)
print("The mean of second split of data for total flow rate (kg/h), containing 1000 points is:", meantot2)
```

```
#Calculating variance of the split data for each flow rates

#For cold water mass flow rate
varcw1 = st.variance(cw1)
varcw2 = st.variance(cw2)

#For hot water mass flow rate
varhw1 = st.variance(hw1)
varhw2 = st.variance(hw2)

#For total mass flow rate
vartot1 = st.variance(tot1)
vartot2 = st.variance(tot2)

#Printing the values
print("The variance of first split of data for flow rate of cold water (kg/h), containing 1500 points is:", varcw1)
print("The variance of second split of data for flow rate of cold water (kg/h), containing 1000 points is:", varcw2)

print("The variance of first split of data for flow rate of hot water (kg/h), containing 1500 points is:", varhw1)
print("The variance of second split of data for flow rate of hot water (kg/h), containing 1000 points is:", varhw2)

print("The variance of first split of data for total flow rate (kg/h), containing 1500 points is:", vartot1)
print("The variance of second split of data for total flow rate (kg/h), containing 1000 points is:", vartot2)
```

Part (C)

Removing outliers now

```
#Function for checking the values for limits between mean plus mins 3 sigma and replacing the ones not in the range with average value of the column
# source: https://stackoverflow.com/questions/33115996/how-do-i-search-a-list-for-a-number-between-two-values
def between(l1,low,high):
    l2 = []
    for i in l1:
        if (i > low and i < high):                      #Checking for the values within the range of mean plus minus 3 sigma
            l2.append(i)                                #Appending the values satisfied to the list l2
        else:
            l2.append(st.mean(l1))                      #Replacing the values which are outside the limits with the average mean of the column and appending to the list l2    
               
    return l2
```

```
#Filtering out values in the range of mean plus minus 3 sigma from the two splits for cold water mass flow rates

#Split 1
lcw1 = meancw1 + (3*(np.sqrt(varcw1)))
lcw2 = meancw1 - (3*(np.sqrt(varcw1)))

#Split 2
lcw3 = meancw2 + (3*(np.sqrt(varcw2)))
lcw4 = meancw2 - (3*(np.sqrt(varcw2)))

#Applying the function for split 1
newcw1 = between(cw1, lcw2, lcw1)
newcw1 = np.array(newcw1)

#Applying the function for split 2
newcw2 = between(cw2, lcw4, lcw3)
newcw2 = np.array(newcw2)
```

```
#Filtering out values in the range of mean plus minus 3 sigma from the two splits for hot water mass flow rates

#Split 1
lhw1 = meanhw1 + (3*(np.sqrt(varhw1)))
lhw2 = meanhw1 - (3*(np.sqrt(varhw1)))

#Split 2
lhw3 = meanhw2 + (3*(np.sqrt(varhw2)))
lhw4 = meanhw2 - (3*(np.sqrt(varhw2)))

#Applying the function for split 1
newhw1 = between(hw1, lhw2, lhw1)
newhw1 = np.array(newhw1)

#Applying the function for split 2
newhw2 = between(hw2, lhw4, lhw3)
newhw2 = np.array(newhw2)
```

```
#Filtering out values in the range of mean plus minus 3 sigma from the two splits for total mass flow rates

#Split 1
ltot1 = meantot1 + (3*(np.sqrt(vartot1)))
ltot2 = meantot1 - (3*(np.sqrt(vartot1)))

#Split 2
ltot3 = meantot2 + (3*(np.sqrt(vartot2)))
ltot4 = meantot2 - (3*(np.sqrt(vartot2)))

#Applying the function for split 1
newtot1 = between(tot1, ltot2, ltot1)
newtot1 = np.array(newtot1)

#Applying the function for split 2
newtot2 = between(tot2, ltot4, ltot3)
newtot2 = np.array(newtot2)
```

Kindly note- The graphs for both splits have been plotted in a single graph, and as the split have omitted few points, hence there is a visible gap between the plots of two splits.

```
#Plotting split 1 & 2 of total mass flow rate
plt.figure(figsize=(20,10)) 
plt.plot(t[0:1500], newtot1, label = 'Total mass flow rate Split 1')                          #Taking the same time split as split 1 for total mass flow rate             
plt.plot(t[1990:], newtot2, label = 'Total mass flow rate Split 2')                           #Taking the same time split as split 2 for total mass flow rate              
plt.xlabel('Time (Sec.)', fontsize=12)
plt.ylabel('Total mass flow rates (split 1 & split 2) (kg/h)',fontsize=12)
plt.title('Total mass flow rates V/s Time plot',fontsize=18)
plt.legend()
```

Part (D)

Mass balance closure error (as given in the document)

Em = mt- mhw- mcw

```
#Calculating mass balance closure errors 

#For split 1
Em_split1 = tot1-hw1-cw1

#For split 2
Em_split2 = tot2-hw2-cw2
```

```
#Mean and variances of the mass balance closure errors

#For split 1
meanem1 = st.mean(Em_split1)
varem1 = st.variance(Em_split1)
print("The mean of the mass balance closure error for split 1 is:", meanem1)
print("The variance of the mass closure balance error fo split 1 is:", varem1)

#For split 2
meanem2 = st.mean(Em_split2)
varem2 = st.variance(Em_split2)
print("The mean of the mass balance closure error for split 2 is:", meanem2)
print("The variance of the mass closure balance error fo split 2 is:", varem2)
```

```
#Plotting histogram for mass balance closure error for split 1
plt.figure(figsize=(15,10)) 
sns.distplot(Em_split1, hist=True, kde=False, 
             bins=int(180/5), color = 'blue',
             hist_kws={'edgecolor':'black'})
# Add labels
plt.title('Histogram showing the mass balance closure error for split 1',fontsize=18)
plt.xlabel('Error Value',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
```

```
#Plotting histogram for mass balance closure error for split 2
plt.figure(figsize=(15,10)) 
sns.distplot(Em_split2, hist=True, kde=False, 
             bins=int(180/5), color = 'blue',
             hist_kws={'edgecolor':'black'})
# Add labels
plt.title('Histogram showing the mass balance closure error for split 2')
plt.xlabel('Error Value')
plt.ylabel('Frequency')
```

####Comment as to why the variance of one of the temperature measurements is different than the other.

The variance for T4 measurements is different than that of T5 measurements as it is evident that the T4 temperature sensor is placed just after the mixing of hot and cold water streams, when the temperature is still fluctuating and hasn't reached a comparitively steady state whereas by the time a reading is recorded in T5 temperature sensor, it has attained a relatively steady state in terms of temperature and the changes are slowly minimized.
